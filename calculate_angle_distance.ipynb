# Calculate Angle

```python
from math import atan2, degrees
def calculate_angle_anticlockwise(p1, p2):
    xDiff = p1[0] - p2[0]
    yDiff = p1[1] - p2[1]
    ang = degrees(atan2(yDiff, xDiff))
    if ang < 0:
        ang += 360
    return ang
```

# Calculate Angle with three points

```python
from math import atan2, degrees
def calculate_angle(p1, p2, p3): #p1 is the middle point
    double_result = atan2(p3[1] - p1[1], p3[0] - p1[0]) - atan2(p2[1] - p1[1], p2[0] - p1[0]);
    ang = degrees(double_result)
    if ang < 0:
        ang += 360
    #if ang > 180:
    #    ang -=180
    return ang
```

# Calculate Distance

```python
from math import atan2, degrees
def calculate_distance(p1, p2):
    xDiff = p2[0] - p1[0]
    yDiff = p2[1] - p1[1]
    distance = np.sqrt(xDiff**2 + yDiff**2)
    return distance
```
